3 - Tire a porcentagem do AGE pela quantidade total de linhas do arquivo.

Temos 2 métodos para a limpeza do arquivo.

1. comando is null = verifica quantos valoresnulos possui no arquivo.

2. comando isna = conta a quantidade de valores não atribuídos/preenchidos no arquivo.

```python
titanic_df.isnull()
```

```python
# Para contar os dados, acrescentamos a soma:

titanic_df.isnull().sum()
```

```python
# Resolução do exercício.

# Pegamos primeiro os valores nulos (isnull).

# Após, multiplicamos por 100 e dividimos pela quantidade total de linhas => [0].

# Delimitamos a qantidade de casas decimais após a vírgula com o comando round.

round((titanic_df.isnull().sum() * 100) / titanic_df.shape[0],2)
```
