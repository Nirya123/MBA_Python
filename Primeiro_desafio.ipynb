- Para cada linha de conversa criada ou todo o arquivo, ele dá o comando <font color="red">split</font>, dividindo as palavras que está lá dentro.
- Após, imprimir (<font color="red">print</font>) a conversa.

```python
from datetime import datetime

abrir_conversa = open('C:/Users/SAMSUNG/Desktop/MBA/zap.txt','r+', encoding = 'utf-8')

for linha in abrir_conversa:
    valores = linha.split()
    print(linha) 
```
